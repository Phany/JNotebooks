TRC02U21 - Voltage probe

50MSa
TRC01I2Pico2 - Picolog probe (TA326) around the insolator that connects to the PCB 
TRC02I22 - PCB current

10MSa
TRC01_I2_Pico_3 - CH1 - Picolog probe (TA326) around the insolator that connects to the PCB 
TRC02_I2_3 - CH2 - PCB current

```python
## import libraries needed 
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
#lOAD DATA
TRC01I2Pico1 = 'D:\d_o_c\Dropbox\Stephanie and me\Coppens\Measurements\MeasC_20210712\Osc_2\TRC01_I2_Pico_1.txt'
TRC01I2Pico2 = 'D:\d_o_c\Dropbox\Stephanie and me\Coppens\Measurements\MeasC_20210712\Osc_2\TRC01_I2_Pico_2.txt'
TRC01I2Pico3 = 'D:\d_o_c\Dropbox\Stephanie and me\Coppens\Measurements\MeasC_20210712\Osc_2\TRC01_I2_Pico_3.txt'
TRC02I22 = 'D:\d_o_c\Dropbox\Stephanie and me\Coppens\Measurements\MeasC_20210712\Osc_2\TRC02_I2_2.txt'
TRC02I23 = 'D:\d_o_c\Dropbox\Stephanie and me\Coppens\Measurements\MeasC_20210712\Osc_2\TRC02_I2_3.txt'
TRC02U21 = 'D:\d_o_c\Dropbox\Stephanie and me\Coppens\Measurements\MeasC_20210712\Osc_2\TRC02_U2_1.txt'
```

```python
# initialise data of lists.
data1 = pd.read_csv(TRC01I2Pico1, sep=",", header=None).transpose()
data2 = pd.read_csv(TRC01I2Pico2, sep=",", header=None).transpose()
data3 = pd.read_csv(TRC01I2Pico3, sep=",", header=None).transpose()
data4 = pd.read_csv(TRC02I22, sep=",", header=None).transpose()
data5 = pd.read_csv(TRC02I23, sep=",", header=None).transpose()
data6 = pd.read_csv(TRC02U21, sep=",", header=None).transpose()
```

```python
df = pd.DataFrame(data1)
df['TRC01I2Pico2'] = data2
df['TRC01I2Pico3'] = data3
df['TRC02I22'] = data4
df['TRC02I23'] = data5
df['TRC02U21'] = data6
df.rename(columns={0 :'TRC01I2Pico1'}, inplace=True )
```

```python
df['num'] = df.index + 1
df['time_us_10MSa'] = df['num']*0.1
df['time_us_50MSa'] = df['num']*0.02
```

The column num is just the number of points, it has no relation with time

```python
fig = px.line(df, x='num', y=['TRC01I2Pico1', 'TRC02U21'], title='I2_Pico & Voltage')
fig.show()
```

```python
fig = px.line(df, x='time_us_50MSa', y=['TRC01I2Pico2', 'TRC02I22'], title='50MSa: I2_Pico & I2_PCB')
fig.show()
```
